# Making a basic figure with matplotlib

```python
# load libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy 
import PIL
import requests
```

```python
# Use a random seed for reproducibility
np.random.seed(613)

# Create a set of random numbers for plotting
x = np.arange(50)
y = np.random.randint(0, 100, 50)

fig, ax = plt.subplots(figsize = (5,3))
ax.scatter(x, y) # plot data as scatter plot
```

```python
fig, ax = plt.subplots(figsize = (5,3))
ax.bar(x,y) # plot data as a bar chart
```

```python
fig, ax = plt.subplots(figsize = (5,3))
ax.barh(x,y) # plot data as horizontal bar chart
```

```python
fig, ax = plt.subplots(figsize = (5,3))
ax.plot(x,y) # plot data as line chart
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.hist(y) # plot data as histograms, for this type of graphs, we only need the y-values
```

```python
fig, ax = plt.subplots(figsize = (5,3))
ax.plot(x,y)

ax.set_title('Total growth over time')
ax.set_ylabel('Total growth')
ax.set_xlabel('Years since start')

fig.tight_layout
```

```python
font1 = {'family':'sans-serif','color':'blue','size':20}

font2 = {'family':'monospace','color':'green',
         'size':14}
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y)

ax.set_title('Total growth over time lets make this a really long title', fontdict = font1, loc='left')
ax.set_ylabel('Total growth', fontdict = font2)
ax.set_xlabel('Years since start', fontdict = font2)
fig.tight_layout()
```

# modifying our data points

```python
fig, ax = plt.subplots(figsize = (5,3))

# customizing the marker format
ax.scatter(x, y,
           marker = '*',
           color = 'indigo')
```

```python
fig, ax = plt.subplots(figsize=(5,3))

ax.plot(x,y,
        marker = '*',
        color = '#7425b9',
        linestyle = '--',
        linewidth = 2)
```

```python
fig, ax = plt.subplots(figsize=(5,3))

ax.plot(x,y,marker = '*',
        markersize = 12,
        color = '#7425b9',
        linestyle = '--',
        linewidth = 2,
        markeredgecolor = '#fa9359',
        markerfacecolor = '#000000')

ax.grid(axis = 'y', color='blue', 
        linewidth = 2, linestyle = '-.')
```
